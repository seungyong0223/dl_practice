```python
# Load modules 
import torch 
import numpy 
```

```python
# Initializing a Tensor 

# From data 
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(f"Tensor 1: \n {x_data} \n")

# From a Numpy array
np_array = np.array(data) 
x_np = torch.from_numpy(np_array)
print(f"Tensor 2: \n {x_np} \n")

# From another tensor
x_ones = torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones} \n")
 
x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# Tensors with random or constant values

shape = (2, 3)  # a tuple of tensor dimensions 
rand_tensor = torch.rand(shape) 
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random tensor: \n {rand_tensor} \n")
print(f"Ones tensor: \n {ones_tensor} \n")
print(f"Zeros tensor: \n {zeros_tensor} \n")
```

```python
# Attributes of a Tensor 
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# Operations on Tensors 

# Move tensors to the accelerator 
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else 'cpu'
tensor = tensor.to(device)
print(f"Device tensor is stored on: {tensor.device}")

# Numpy-like indexing and slicing
tensor = torch.ones(4, 4) 
tensor[:,1] = 0
print(f"Indexing and Slicing: \n {tensor} \n") 

# Joining 
t1 = torch.cat([tensor, tensor, tensor], dim=1) 
print(f"Joining: \n {t1} \n") 

# Element-wise product
print(f"tensor.mul(tensor): \n {tensor.mul(tensor)} \n")
print(f"tensor * tensor: \n {tensor * tensor} \n")

# Matrix multiplication 
print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)} \n")
print(f"tensor @ tensor.T \n {tensor @ tensor.T} \n")
```

```python
# In-place operations 
# Note that it is problematic when computing gradients 
print(tensor, "\n")
tensor.add_(5)
print(tensor, "\n")
```

```python
# Tensor & NumPy 
# Tensors on the CPU and NumPy arrays can share their memory locations. 
# Changing one will change the oher. 

# Tensor to NumPy array
t = torch.ones(5) 
n = t.numpy() 
print("\n Tensors to NumPy \n")
print(f"t: {t}")
print(f"n: {n}")

# add 
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")

# NumPy array to Tensor 
n = np.ones(5) 
t = torch.from_numpy(n)
print("\n NumPy to Tensor \n")
print(f"t: {t}")
print(f"n: {n}")

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
